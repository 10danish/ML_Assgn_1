```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#I have merged the both csv files(x&y) into one csv file for ease.
data=pd.read_csv('data_xy.csv')
X=data['X-values'].values
Y=data['Y-values'].values
th0=0
th1=0
alpha=0.5
epochs=10000
m=len(X)

cost=np.array([])
ep=np.array([])
X=(X-np.mean(X))/np.std(X)
def cost_function(th0,th1,Y,X):
    return np.sum(np.square(th0+th1*X-Y))/(2*m)

for epoch in range(epochs):
    pred=th0+th1*X
    error=pred-Y
    th0_gradient=(1/m) * np.sum(error)    
    th1_gradient=(1/m) * np.sum(error * X)
    
    th0 -= alpha * th0_gradient
    th1 -= alpha * th1_gradient
    
    cost=np.append(cost,cost_function(th0, th1,Y, X))
    
pred_Y=th0+th1*X
# print(th0+th1*3)
ep = np.arange(epochs)
plt.scatter(X,Y, label='Original Data')
plt.plot(X, pred_Y, color='red', label='Linear Regression Line')
# plt.ylim(0.98, 1.003)
plt.title('Linear Regression with Mean Squared Error')
plt.xlabel('X-values')
plt.ylabel('Y-values')
plt.legend()
plt.show()

print(f"Theta0: {th0}, Theta1: {th1}")
final_mse = cost_function(th0,th1,Y,X)
print(f"Final MSE: {final_mse}")

plt.scatter(ep,cost,label='Cost functions values')
plt.title('Cost function vs Iteration(batch gradient)')
plt.xlabel('Iterations')
plt.ylabel('cost values')
plt.legend()
plt.show()
# print(ep.size)
# print(cost.size)
```

# Using stochastic gradient descent

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#I have merged the both csv files(x&y) into one csv file for ease.
data=pd.read_csv('data_xy.csv')
X=data['X-values'].values
Y=data['Y-values'].values
th0=0
th1=0
alpha=0.001
epochs=10000
m=len(X)

# cost=np.array([])
ep=np.array([])
X=(X-np.mean(X))/np.std(X)
def cost_function(th0,th1,Y,X):
    return np.sum(np.square(th0+th1*X-Y))/(2*m)

def stochastic_gradient_descent(th0, th1, X, Y, alpha, epochs):
    cost_vals = []

    for epoch in range(epochs):
        for i in range(len(X)):
            pred = th0 + th1 * X[i]
            error = pred - Y[i]

            th0 -= alpha * error
            th1 -= alpha * error * X[i]

        cost_vals.append(cost_function(th0, th1, Y, X))

    return th0, th1, cost_vals


th0, th1, cost = stochastic_gradient_descent(th0, th1, X, Y, alpha, epochs)
pred_Y=th0+th1*X
# print(th0+th1*3)
ep = np.arange(epochs)
plt.scatter(X,Y, label='Original Data')
plt.plot(X, pred_Y, color='red', label='Linear Regression Line')
# plt.ylim(0.98, 1.003)
plt.title('Linear Regression with Mean Squared Error')
plt.xlabel('X-values')
plt.ylabel('Y-values')
plt.legend()
plt.show()

print(f"Theta0: {th0}, Theta1: {th1}")
final_mse = cost_function(th0,th1,Y,X)
print(f"Final MSE: {final_mse}")
# Plotting the cost function versus iteration for SGD
plt.scatter(ep,cost,label='Cost functions values')
plt.title('Cost function vs Iteration (stochastic)')
plt.xlabel('Iteration')
plt.ylabel('cost values')
plt.legend()
plt.show()
# print(ep.size)
# print(len(cost))
```

# Using mini-batch gradient descent

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#I have merged the both csv files(x&y) into one csv file for ease.
data=pd.read_csv('data_xy.csv')
X=data['X-values'].values
Y=data['Y-values'].values
th0=0
th1=0
alpha=0.001
epochs=10000
batch_size=20
m=len(X)

# cost=np.array([])
ep=np.array([])
X=(X-np.mean(X))/np.std(X)
def cost_function(th0,th1,Y,X):
    return np.sum(np.square(th0+th1*X-Y))/(2*m)

def mini_batch_gradient_descent(th0, th1, X, Y, alpha, epochs, batch_size):
    cost_vals = []

    for epoch in range(epochs):
        for i in range(0, m, batch_size):
            X_batch = X[i:i+batch_size]
            Y_batch = Y[i:i+batch_size]

            pred = th0 + th1 * X_batch
            error = pred - Y_batch

            th0 -= alpha * np.sum(error) / batch_size
            th1 -= alpha * np.sum(error * X_batch) / batch_size

        cost_vals.append(cost_function(th0, th1, Y, X))

    return th0, th1, cost_vals


th0, th1, cost = mini_batch_gradient_descent(th0, th1, X, Y, alpha, epochs, batch_size)
pred_Y=th0+th1*X
# print(th0+th1*3)
ep = np.arange(epochs)
plt.scatter(X,Y, label='Original Data')
plt.plot(X, pred_Y, color='red', label='Linear Regression Line')
# plt.ylim(0.98, 1.003)
plt.title('Linear Regression with Mean Squared Error')
plt.xlabel('X-values')
plt.ylabel('Y-values')
plt.legend()
plt.show()

print(f"Theta0: {th0}, Theta1: {th1}")
final_mse = cost_function(th0,th1,Y,X)
print(f"Final MSE: {final_mse}")
# Plotting the cost function versus iteration for SGD
plt.scatter(ep,cost,label='Cost functions values')
plt.title('Cost function vs Iteration (mini-batch)')
plt.xlabel('Iterations')
plt.ylabel('cost values')
plt.legend()
plt.show()
# print(ep.size)
# print(len(cost))
```
